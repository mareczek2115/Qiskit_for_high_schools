### Are all qubits equally good?
Nowadays qubits are far from being perfect. So far, research is ongoing to improve the quantum unit, which will allow achieving a quantum advantage (which we will describe this later). For now, we can assume that one perfect logical qubit is equal to approximately 50 qubits.

The biggest problem with qubits is decoherence. It is based on the fact that quantum states are extremely unstable. Under the influence of external factors, the system "drops" from the superposition state and "jumps" to one of the stationary states. This is done in a tiny fraction of a second. Even the smallest contact with the surroundings can affect the measurement result (which is one of the reasons why the temperature must be so low).

### Quantum Error Correction
To protect quantum information from errors due to decoherence and other quantum noise we use Quantum Error Correction (QEC). Quantum error correction is essential if one is to achieve fault-tolerant quantum computation that can deal not only with noise on stored quantum information but also with faulty quantum gates and faulty measurements. Classic error correction uses redundancy. The easiest way is to store information multiple times, and if it later turns out that the copies do not match, take the majority of the votes. Let's present it with an example:

Suppose we have three bits. The error is that one of these bits is zero and the other bits are ones. If we assume that the errors are independent and occur with a certain probability *p*, it is most likely a one-bit error and the transmitted message has three ones.

### Backend map
In Qiskit, there is easy way to check map of qubits implemented in specified backend. In order for us to be able to use IBM computers, we must authenticate our devices with a special token, which is available on the [IBM Quantum Experience](https://quantum-computing.ibm.com/) website. For this purpose, we recommend using the [official Qiskit guide](https://www.youtube.com/watch?v=M4EkW4VwhcI) (second half of the video). Now let's get back to our exercise:

```python
from qiskit import *
provider = IBMQ.load_account()
backend = provider.get_backend('ibmq_valencia')
import pprint
pprint.pprint(backend.configuration().coupling_map)
```

As you can see above, first thing we got is array of some pairs. These are direct connections between qubits, as can be seen in the diagram below:

```python
from qiskit.tools.visualization import plot_error_map
plot_error_map(backend)
```

As you can see on plot, there is map with qubits connections which complies with previous array of connections. In addition, the probability of an error on a particular qubit is shown. The lighter the color the more likely it is. The graph on the left shows the error probability of each qubit. As you can see, almost all qubits except 4th have a 3% error probability. The error probability in one-qubit gates is presented below from the left, on the example of a Hadamard gate. The highest chance of error is on the 3rd qubit at 0.064%. The last one is the error probability on the two qubit gate on the example of CNOT. We read data in the same way as indirectly, we only look at connections between qubits.
